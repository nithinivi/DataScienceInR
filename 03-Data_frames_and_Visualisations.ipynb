## Grouping and Facets

````r
df <- data.frame(
  x = rnorm(120, c(0, 2, 4)),
  y = rnorm(120, c(1, 2, 1)),
  z = letters[1:3]
)
head(df)
````

````r
library(repr)
options(repr.plot.width=5, repr.plot.height=2)
````

````r
ggplot(df, aes(x, y)) + 
  geom_point(aes(colour = z))
````

Faceting is an alternative to using aesthetics (like colour, shape or size) to differentiate groups. Both techniques have strengths and weaknesses, based around the relative positions of the subsets. With faceting, each group is quite far apart in its own panel, and there is no overlap between the groups. .This is good if the groups overlap a lot, but it does make small differences harder to see. When using aesthetics to differentiate groups, the groups are close together and may overlap, but small differences are easier to see.

````r
ggplot(df, aes(x, y)) + 
  geom_point() + 
  facet_wrap(~z)
````

Comparisons between facets often benefit from some thoughtful annotation. For example, in this case we could show the mean of each group in every panel

````r
df_sum <- df %>% 
  group_by(z) %>% 
  summarise(x = mean(x), y = mean(y)) %>%
  rename(z2 = z)

head(df_sum)
````

````r
ggplot(df, aes(x, y)) + 
  geom_point(alpha=.4) + 
  geom_point(data = df_sum, aes(colour = z2), size = 3) + 
  facet_wrap(~z)
````

````r
df2 <- select(df, -z)
head(df2)
````

````r


ggplot(df, aes(x, y)) + 
  geom_point(data = df2, colour = "grey70") +
  geom_point(aes(colour = z)) +
  facet_wrap(~z)
````
